```python
from bs4 import BeautifulSoup

with open("your_posts_1.html", encoding="utf8") as page:
    soup = BeautifulSoup(page, 'html.parser')
```

```python
soup
```

```python
contents = soup.find_all("div",class_="_2pin")
```

```python
x=[]
for i in contents:
    x.append(i.get_text())
```

```python
x
```

```python
y = ' '.join(x)
```

```python
y
```

```python
y=y.replace("PM","")
y=y.replace("Jan","")
y=y.replace("Feb","")
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import random
 
def grey_color_func(word, font_size, position, orientation, random_state=None, **kwarfs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60,100)

from PIL import Image
import numpy as np

mask=np.array(Image.open("t.png"))

wordcloud = WordCloud(background_color="black",
                     max_words=2000,
                     width=1020,
                     height=720,
                     mask=mask)
wordcloud.generate(y)
plt.figure(figsize=(20,8))
plt.title("Facebook WordCloud")
plt.imshow(wordcloud.recolor(color_func=grey_color_func, random_state=3), interpolation='bilinear')
plt.axis('off')
plt.show()
```
